```python
#csv 파일 별로 읽어서 작업
path = '/content/drive/MyDrive/final_project_work/csv'
csv_list = os.listdir(path)
def checking_null(food_cls):
    food_cls = food_cls
    for file_name in csv_list:
        file_path = os.path.join(path, file_name)
        checked_csv = pd.read_csv(file_path, low_memory=False)
        #필수 영양소 columns
        essential_cols = ['단백질(g)','지방(g)','탄수화물(g)']
        for col in essential_cols:
            #필수 영양소 columns에 들어있는 문자열 (공백, 기호) Nan값으로 처리
            checked_csv[col] = checked_csv[col].replace(['-',' ',''],np.nan)
            checked_csv[col] = pd.to_numeric(checked_csv[col], errors='coerce')
        #중앙값 확인
        #중앙값 확인하고 결측치에만 중앙값 기입해야함.
        #실제로 0인 값과 Nan값 구별
        # if (checked_csv["식품대분류명"] == food_cls).all():
        #     for i in essential_cols:
        #         print(checked_csv[i].describe(),"\n\n")

        subset = checked_csv[checked_csv["식품대분류명"] == food_cls].copy()
        if not subset.empty:
            for col in essential_cols:
                median_value = subset[col].describe()['50%']
                nan_mask = subset[col].isnull()
                zero_mask = (subset[col] == 0)
                if nan_mask.sum() > 0:
                    print(f"{col}결측치 확인 : {nan_mask.sum()}개")
                    print(f"0 : {zero_mask.sum()}개", "\n")
                    subset[col] = subset[col].fillna(median_value)
                    #after fillna
                    after_nan = subset[col].isnull()
                    after_zero = (subset[col] == 0)
                    print(f"=={file_name},{col} 최종 결측치 : {after_nan.sum()}개==")
                    print(f"0 : {after_zero.sum()}개", "\n")
                    checked_csv.loc[subset.index, col] = subset[col]

        checked_csv.to_csv(file_path, index=False)
        print(f"저장 완료: {file_path}")










```

```python
food_cls_list = csv['식품대분류명'].unique().tolist()
food_cls_list
for fc in food_cls_list:
    checking_null(fc)
```

```python
import pandas as pd
import os

# 1️⃣ CSV 파일들이 저장된 폴더 경로
path = '/content/drive/MyDrive/final_project_work/csv'

# 2️⃣ 폴더 안의 모든 csv 파일 불러오기
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]

# 3️⃣ 모든 csv 파일을 하나의 DataFrame으로 합치기
merged_csv = pd.concat([pd.read_csv(os.path.join(path, f)) for f in csv_files], ignore_index=True)

# 4️⃣ 합친 파일 저장
merged_csv.to_csv('/content/drive/MyDrive/final_project_work/merged_food.csv', index=False, encoding='utf-8')

print(f"총 {len(csv_files)}개의 파일이 합쳐졌습니다.")
print(f"최종 행 개수: {len(merged_csv)}")
```
